```python
import matplotlib.pyplot as plt
import pandas_datareader as web
import pandas as pd
import yahoo_fin.stock_info as si
```

```python
def get_balance(company):
    return si.get_balance_sheet(company)
```

```python
def graph_ratios(balance_sheet,company):
    df = pd.DataFrame(columns = ["2021","2020","2019","2018"], data=balance_sheet.values, index=balance_sheet.index)
    df=df[["2018","2019","2020","2021"]]
    totalDebt=df.loc["shortLongTermDebt"]+df.loc["longTermDebt"]
    df.loc["totalDebt"]=totalDebt
    debttoassets=df.loc["totalDebt"]/df.loc["totalAssets"]
    df.loc["debttoassets"]=debttoassets

    currentratio=df.loc["totalCurrentAssets"]/df.loc["totalCurrentLiabilities"]
    df.loc["currentratio"]=currentratio
    cashratio=df.loc["cash"]/df.loc["totalCurrentLiabilities"]
    df.loc["cashratio"]=cashratio
    debttoequity=df.loc["totalDebt"]/df.loc["totalStockholderEquity"]
    df.loc["debttoequity"]=debttoequity

    f = plt.figure(figsize=(18,5))
    if company not in dict_companies:
        title_company = company
    else:
        title_company=dict_companies[company]
    f.suptitle(title_company)
    ax1 = f.add_subplot(141)
    ax2 = f.add_subplot(142)
    ax3 = f.add_subplot(143)
    ax4 = f.add_subplot(144)

    ax1.plot(df.loc["currentratio"], marker='*')
    ax1.set_title('Current Ratio')
    ax2.plot(df.loc["cashratio"], marker='*')
    ax2.set_title('Cash Ratio')
    ax3.plot(df.loc["debttoassets"], marker='*')
    ax3.set_title('Debt to Assets')
    ax4.plot(df.loc["debttoequity"], marker='*')
    ax4.set_title('Debt to Equity')
    plt.show()
```

```python
list_companies=[
"DPZ",
"PZZA",
"MCD",
"SIX",
"SEAS",
"GDRX",
"HLF"
]

dict_companies={
"DPZ":"Domino's Pizza, Inc",
"HLF":"Herbalife Nutrition Ltd.",
"MCD": "McDonald's Corporation",
"PZZA": "Papa John's International, Inc. (PZZA)",
"SIX": "Six Flags Entertainment Corporation",
"SEAS": "SeaWorld Entertainment, Inc. (SEAS)",
"GDRX": "GoodRx Holdings, Inc."
}
```

```python
#company="TSLA"
company=list_companies[0]
print(company)
balance_sheet=get_balance(company)
balance_sheet
```

```python
graph_ratios(balance_sheet,company)
```
